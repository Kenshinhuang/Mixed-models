```python
url='https://bd29ee0e-54ab-4daa-9671-d153865d1620.usrfiles.com/ugd/bd29ee_57b940b192b3419aafaac279ea9c256b.csv'
df_original=pd.read_csv(url,header=None)
df_original.head()
```

The task for this dataset is clustering. We are going to use PCA to reduce the dimension before clustering. 

Drop the first column. It is just the dates.

```python
df=df_original.drop(columns=0)
df.head()
```

```python
from sklearn.decomposition import PCA
n_components = 10
pca = PCA(n_components=n_components,copy=True,whiten=True,svd_solver="randomized",random_state=0)
df_pca = pd.DataFrame(pca.fit_transform(df))
df_pca.head()
```

PCA has reduced the dimensionality of the dataset from 38 (excluding the date) to 10.

We are going to try the most well-known clustering algorithm KMeans first.

```python
import numpy as np
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


n_clusters=np.arange(2,11)

for n_clusters in n_clusters:
    clusterer = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels = clusterer.fit_predict(df_pca)
    silhouette_avg = silhouette_score(df_pca, cluster_labels,metric='euclidean')
    print("For n_clusters =", n_clusters,
          " ,the average silhouette_score is :", silhouette_avg,)
```

It seems that KMeans is not the right clustering algorithm for this dataset. Let's try Affinity Propagation andd evaluate the result with Davies-Bouldin Index.

```python
from sklearn.cluster import AffinityPropagation
from sklearn.metrics import davies_bouldin_score


clusterer = AffinityPropagation(max_iter=600)
cluster_labels = clusterer.fit_predict(df_pca)
davies_bouldin_score = davies_bouldin_score(df_pca, cluster_labels)
score=silhouette_score(df_pca, cluster_labels,metric='euclidean')
print("The Davies Bouldin score is: ",davies_bouldin_score)
print("The Silhouette score is: ",score)
```

Davies Bouldin Score is above 1.18, while we would want it to be close to 0. 

Mean-shift.

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

bandwidth = estimate_bandwidth(df_pca,quantile=0.95, n_jobs=-1)
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
labels=ms.fit_predict(df_pca)
labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)
print("The number of clusters is: ",n_clusters_)
print("The silhouette score is: ",silhouette_score(df_pca, labels))
```

So far, MeanShift with band width gives us the best result.

Spectral clustering.

```python
from sklearn.cluster import SpectralClustering

n_clusters=np.arange(2,13)

for n_clusters in n_clusters:
    clusterer = SpectralClustering(n_clusters=n_clusters, assign_labels="discretize",random_state=0)
    cluster_labels = clusterer.fit_predict(df_pca)
    silhouette_avg = silhouette_score(df_pca, cluster_labels,metric='euclidean')
    print("For n_clusters =", n_clusters,
          " ,the average silhouette_score is :", silhouette_avg,)
```

There is some hope here. SpectralClustering with 2 clusters seems to be the best algorithm at this moment.

```python
from sklearn.cluster import AgglomerativeClustering

n_clusters=np.arange(2,13)
for n_clusters in n_clusters:
    clusterer = AgglomerativeClustering(n_clusters=n_clusters)
    cluster_labels = clusterer.fit_predict(df_pca)
    silhouette_avg = silhouette_score(df_pca, cluster_labels)
    print("For n_clusters =", n_clusters,
          " ,the average silhouette_score is :", silhouette_avg,)
```

Not as good as SpectralClustering.

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=7).fit(df_pca)
labels = db.labels_
print("The number of clusters is: ",len(np.unique(labels)))
print("Silhouette score: %0.3f" % metrics.silhouette_score(df_pca, labels))
```

Best result so far.

```python
from sklearn.cluster import Birch

clusterer = Birch(threshold=0.9,branching_factor=100,n_clusters=2)
cluster_labels = clusterer.fit_predict(df_pca)
silhouette_avg = silhouette_score(df_pca, cluster_labels,metric='euclidean')
print("For n=2, the average silhouette_score is :", silhouette_avg,)

clusterer = Birch(threshold=0.9,branching_factor=100,n_clusters=3)
cluster_labels = clusterer.fit_predict(df_pca)
silhouette_avg = silhouette_score(df_pca, cluster_labels,metric='euclidean')
print("For n=3, the average silhouette_score is :", silhouette_avg,)

clusterer = Birch(threshold=0.9,branching_factor=100,n_clusters=4)
cluster_labels = clusterer.fit_predict(df_pca)
silhouette_avg = silhouette_score(df_pca, cluster_labels,metric='euclidean')
print("For n=4, the average silhouette_score is :", silhouette_avg,)

clusterer = Birch(threshold=0.9,branching_factor=100,n_clusters=5)
cluster_labels = clusterer.fit_predict(df_pca)
silhouette_avg = silhouette_score(df_pca, cluster_labels,metric='euclidean')
print("For n=5, the average silhouette_score is :", silhouette_avg,)
```

We have tried all available SKlearn clustering algorithms. **For this particular dataset and with the dimensionality of the dataset reduced to 10 by PCA, DBSCAN with eps=0.7 gives the best result.**

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import davies_bouldin_score

db = DBSCAN(eps=7).fit(df_pca)
labels = db.labels_
print("The number of clusters is: ",len(np.unique(labels)))
print("Silhouette score is: %0.3f" % metrics.silhouette_score(df_pca, labels))
print("The Davies Bouldin score is: ",davies_bouldin_score(df_pca, labels))
```
